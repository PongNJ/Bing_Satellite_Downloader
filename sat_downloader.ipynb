## Calculating Bounding Box Coordinates for a Given Lat/Lon in WGS84
This script calculates the top-left (left-up) and bottom-right (right-down) corner coordinates of a 200m x 200m bounding box in the WGS84 coordinate system.

Given
- A central point specified by its latitude and longitude in WGS84 format.
- A square box size of 200 meters.
The goal is to compute:

Output
- The left-up corner coordinate (latitude, longitude).
- The right-down corner coordinate (latitude, longitude).


### Prepare Bounding box to retrive Satellite imagery

```python
# Apply the function to calculate bounding boxes for each row with a box size of 200 meters
df2["BoundingBox"] = df2.apply(lambda row: calculate_bounding_box(row["Latitude start"], row["Longitude start"], box_size_meters=200), axis=1)

# Separate the results into columns for AerialImageRetrieval
df2[["Lat1", "Lon1", "Lat2", "Lon2"]] = pd.DataFrame(df2["BoundingBox"].tolist(), index=df2.index)

# Drop the intermediate BoundingBox column if not needed
df2 = df2.drop(columns=["BoundingBox"])

# Print the final dataframe
print(df2) # so the df will have needed columns: Lat1, Lon1, Lat2, Lon2 to use for aerial image retrieval
```

### Retrive images using index
I used index as folder and named image with zoom level and timestamp here because this can use to implement later when we can get termporal satellite data or other APIs


```python
import os
from datetime import datetime

# Ensure output folder exists
output_folder = "output"
os.makedirs(output_folder, exist_ok=True)

# Keep track of processed indices to avoid redundant downloads
processed_indices = set()

# Loop through each unique index in the DataFrame
for index, row in df2.iterrows():
    current_index = row["index"]  # Unique identifier for folder naming

    # Skip if the index is already processed
    if current_index in processed_indices:
        continue
    processed_indices.add(current_index)

    # Extract bounding box coordinates for this row
    lat1, lon1, lat2, lon2 = row["Lat1"], row["Lon1"], row["Lat2"], row["Lon2"]

    # Create a subfolder for the current index
    subfolder = os.path.join(output_folder, str(current_index))
    os.makedirs(subfolder, exist_ok=True)

    # Create an instance of AerialImageRetrieval for this bounding box
    imgretrieval = AerialImageRetrieval(lat1, lon1, lat2, lon2)
    print(f"Processing index {current_index}...")

    # Update the target folder for the AerialImageRetrieval instance
    imgretrieval.tgtfolder = subfolder  # Set the target folder to the subfolder

    # Retrieve the image and save it in the subfolder
    if imgretrieval.max_resolution_imagery_retrieval():
        print(f"Image for index {current_index} saved in {subfolder}.")
    else:
        print(f"Failed to retrieve image for index {current_index}.")
```
